```python

def plot_co2(sensores, fecha1="2021-09-23 13:20", deltat="1H",
             lista=['fluke',"s8lp","scd30","sen0220"],
             ymin=300, ymax=500, calibracion=False, savefig=False):
    fig, ax = plt.subplots(figsize=(7, 2), constrained_layout=True)

    fecha1 = parse(fecha1)
    fecha2 = fecha1 + pd.Timedelta(deltat)

    color = {"s8lp": "b", "sen0220": "g", "scd30": 'r', 'fluke': 'k-'}
    for sensor in lista:
        if sensor == "fluke":
            etiqueta = "Fluke"
        else:
            etiqueta = sensor.upper()
        ax.plot(sensores[sensor], color[sensor], label=etiqueta)
    
    # ax.fill_between(sensores.fluke.index, sensores.fluke + 50, sensores.fluke - 50, alpha=0.2, color='k')

    ax.set_ylim(ymin, ymax)
    ax.set_ylabel("CO2 [ppm]")
    ax.set_xlabel("Hora")
    ax.set_yticks(np.arange(ymin - 50, ymax + 50, 100))
    
    # Configurar formato del eje x
    ax.xaxis.set_major_locator(mdates.HourLocator(interval=3))  # Ticks cada 3 horas
    ax.xaxis.set_major_formatter(mdates.ConciseDateFormatter(mdates.AutoDateLocator()))

    ax.set_xlim(fecha1, fecha2)
    ax.grid()
    ax.legend()

    if savefig:
        fig.savefig("../img/history.png", dpi=300)
        print("Se guard√≥ una figura")
    if calibracion:
        return sensores[fecha1:fecha2]
```

```python
fecha1 = "2021-09-23 11:20"
ymin = 330
ymax = 500
plot_co2(datos,fecha1=fecha1,deltat="400Min",ymin=ymin,ymax=ymax)
calibra430 = plot_co2(datos,fecha1="2021-09-23 13:20",deltat="40Min",
                      ymin=ymin,ymax=ymax,
                      calibracion=True)
```

```python
plot_co2(datos,fecha1="2021-09-29 14:30",deltat="30H",ymin=400,ymax=800)
calibra650 = plot_co2(datos,fecha1="2021-09-29 20:00",deltat="1H",ymin=650,ymax=800,calibracion=True,savefig=True)
```

```python
calibra = pd.concat([calibra430,calibra650,calibra850,calibra1150],axis=0)
```
